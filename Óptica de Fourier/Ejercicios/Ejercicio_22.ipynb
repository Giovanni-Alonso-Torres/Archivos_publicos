```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-4, 4, 1000)
```

```python
Λ = (1-abs(x))*(abs(x)<1)
```

```python
plt.figure()
plt.plot(x, Λ)
plt.show()
```

```python
f = np.linspace(-4, 4, 1000)
ftΛ = (np.sin(np.pi*f)/(np.pi*f))**2
```

```python
plt.figure()
plt.plot(f, abs(ftΛ), color='red')
plt.show()
```
