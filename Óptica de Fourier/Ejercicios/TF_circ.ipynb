```python
from scipy.special import jv
import numpy as np
import matplotlib.pyplot as plt
```

```python
ρ = np.linspace(-3, 3, 10000)
bsinc = jv(1, 2*np.pi*ρ)/ρ
sinc = np.sin(2*np.pi*ρ)/(2*ρ)
```

```python
fig = plt.figure(figsize=(12,9))
ax = fig.add_subplot(111)

ax.plot(ρ, bsinc, color='tab:orange')
ax.plot(ρ, sinc, '--', color='tab:grey')
ax.set_xlabel(r'$\rho$', fontsize=20)
ax.set_ylabel(r'$\mathcal{F}\{\mathrm{circ}\}$', fontsize=20)
ax.set_xlim(ρ[0], ρ[-1])
plt.savefig('../TF_circ.png', dpi=150, format='png', transparent=True)
plt.show()
```
