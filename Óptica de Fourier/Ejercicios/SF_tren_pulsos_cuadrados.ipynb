```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-12, 12, 1000000)
g = [4 if np.abs(a) < 2 or np.abs(a-8) < 2 or np.abs(a+8) < 2 else 0 for a in x]
g10 = 2 + 8/np.pi * sum(((-1)**m/(2*m+1))*np.cos((2*m+1)*np.pi/4*x) for m in range(10))
g100 = 2 + 8/np.pi * sum(((-1)**m/(2*m+1))*np.cos((2*m+1)*np.pi/4*x) for m in range(100))
g1000 = 2 + 8/np.pi * sum(((-1)**m/(2*m+1))*np.cos((2*m+1)*np.pi/4*x) for m in range(1000))
```

```python
fig = plt.figure(figsize=(12,9))
ax = fig.add_subplot(111)
inset = fig.add_axes([0.45, 0.65, 0.4, 0.2])
#left, bottom, width, height

ax.plot(x, g, color='grey')
ax.plot(x, g10, color='tab:blue')
inset.plot(x, g, color='grey')
inset.plot(x, g10, color='tab:blue')
ax.set_xlabel(r'$x$', fontsize=20)
ax.set_ylabel(r'g(x)', fontsize=20)
ax.set_xlim(x[0], x[-1])
ax.set_ylim(-1,8)
inset.set_xlim(1.5, 2.01)
inset.set_ylim(3.5,4.5)
plt.text(0.03, 0.9, r'10 términos, $T=8$', fontsize = 20, transform=ax.transAxes)
plt.savefig('../01_10_terminos.png', dpi=150, format='png', transparent=True)
plt.show()
```

```python
fig = plt.figure(figsize=(12,9))
ax = fig.add_subplot(111)
inset = fig.add_axes([0.45, 0.65, 0.4, 0.2])
#left, bottom, width, height

ax.plot(x, g, color='grey')
ax.plot(x, g100, color='tab:green')
inset.plot(x, g, color='grey')
inset.plot(x, g100, color='tab:green')
ax.set_xlabel(r'$x$', fontsize=20)
ax.set_ylabel(r'g(x)', fontsize=20)
ax.set_xlim(x[0], x[-1])
ax.set_ylim(-1,8)
inset.set_xlim(1.5, 2.01)
inset.set_ylim(3.5,4.5)
plt.text(0.03, 0.9, r'100 términos, $T=8$', fontsize = 20, transform=ax.transAxes)
plt.savefig('../01_100_terminos.png', dpi=150, format='png', transparent=True)
plt.show()
```

```python
fig = plt.figure(figsize=(12,9))
ax = fig.add_subplot(111)
inset = fig.add_axes([0.45, 0.65, 0.4, 0.2])
#left, bottom, width, height

ax.plot(x, g, color='grey')
ax.plot(x, g1000, color='tab:red')
inset.plot(x, g, color='grey')
inset.plot(x, g1000, color='tab:red')
ax.set_xlabel(r'$x$', fontsize=20)
ax.set_ylabel(r'g(x)', fontsize=20)
ax.set_xlim(x[0], x[-1])
ax.set_ylim(-1,8)
inset.set_xlim(1.95, 2.01)
inset.set_ylim(3.5,4.5)
plt.text(0.03, 0.9, r'1000 términos, $T=8$', fontsize = 20, transform=ax.transAxes)
plt.savefig('../01_1000_terminos.png', dpi=150, format='png', transparent=True)
plt.show()
```

```python
fig = plt.figure(figsize=(15,10))
ax = fig.add_subplot(111)

ax.plot(x, g, color='grey')
ax.plot(x, g10, color='tab:blue', label='10 términos')
ax.plot(x, g100, color='tab:green', label='100 términos')
ax.plot(x, g1000, color='tab:red', label='1000 términos')
ax.set_xlabel(r'$x$', fontsize=20)
ax.set_ylabel(r'g(x)', fontsize=20)
ax.set_xlim(1.5, 2.003)
ax.set_ylim(3.5,4.5)
plt.text(0.03, 0.9, r'$T=8$', fontsize = 20, transform=ax.transAxes)
plt.legend(loc='best', bbox_to_anchor=(0.01, 0.65), fontsize=20)
plt.savefig('../01_todos.png', dpi=150, format='png', transparent=True)
plt.show()
```
