```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, ifft, fftfreq, fftshift, ifftshift
```

```python
N = 1000
m = 0.5
#λ = 500e-9
#L = 10*λ
Λ = 1/10
χ = np.linspace(-5, 5, N)
Δχ = χ[1]-χ[0]
ζ = np.linspace(20000, 20025, N)
X, Z = np.meshgrid(χ, ζ)
Ia = abs(1+m*np.exp(-1j*np.pi*Λ*Z)*np.cos(2*np.pi*X))**2/4
```

```python
plt.figure(figsize=(10,8))
plt.pcolormesh(Z, X, Ia, shading='auto')
plt.show()
```

```python
t = (1+m*np.cos(2*np.pi*χ))/2
fftt = fftshift(fft(t))
fx = fftshift(fftfreq(N, Δχ))
Hf = []
for z in ζ:
    Hf.append(np.exp(-1j*np.pi*Λ*z*fx**2))
If = []
for i in range(len(ζ)):
    If.append(abs(ifftshift(ifft(fftt*Hf[i])))**2)
#Uf = np.array(Uf)
```

```python
plt.figure(figsize=(10,8))
plt.pcolormesh(Z, X, If, shading='auto')
plt.show()
```
