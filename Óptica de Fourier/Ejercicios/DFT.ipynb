```python
from scipy.special import jv
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
def circ(X, Y):
    data = []
    for y in Y[::-1]:
        subdata = []
        for x in X:
            r = np.sqrt(x**2+y**2)
            value = 1 if r < 1 else 1/2 if r == 1 else 0
            subdata.append(value)
        data.append(subdata)
    return np.array(data)
```

```python
wx = 2*250
wy = 2*250
Nx = 1000
Ny = 1000
Bx = Nx/wx
By = Ny/wy
```

```python
Bx
```

```python
x = np.linspace(-wx/2, wx/2, Nx)
Δx = x[1]-x[0]
y = np.linspace(-wy/2, wy/2, Ny)
Δy = y[1]-y[0]
X, Y = np.meshgrid(x, y)
Z = circ(x, y)
```

```python
fx = np.linspace(-Bx/2, Bx/2, Nx)
fy = np.linspace(-By/2, By/2, Ny)
Fx, Fy = np.meshgrid(fx, fy)
```

```python
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm, linewidth=0, antialiased=False)
```

```python
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
surf = ax.plot_surface(Fx, Fy, abs(np.fft.fftshift(np.fft.fft2(Z))*Δx*Δy)**2, cmap=cm.coolwarm, linewidth=0, antialiased=False)
```

```python
Ρ = np.sqrt(Fx**2+Fy**2)
Fz = jv(1, 2*np.pi*Ρ)/Ρ
```

```python
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
surf = ax.plot_surface(Fx, Fy, abs(Fz)**2, cmap=cm.coolwarm, linewidth=0, antialiased=False)
```

```python
wx = 2*100
wy = 2*100
Nx = 1000
Ny = 1000
Bx = Nx/wx
By = Ny/wy
w = 5
fx = np.linspace(-Bx/2, Bx/2, Nx)
fy = np.linspace(-By/2, By/2, Ny)
Fx, Fy = np.meshgrid(fx, fy)
```

```python
x = np.linspace(-wx/2, wx/2, Nx)
y = np.linspace(-wy/2, wy/2, Ny)
X, Y = np.meshgrid(x, y)
def sen(X, Y):
    data = []
    for y in Y[::-1]:
        subdata = []
        for x in X:
            value = np.sin(2*np.pi*x) if -w/2 < x < w/2 and -w/2 < y < w/2 else 0
            subdata.append(value)
        data.append(subdata)
    return np.array(data)
    
Z = sen(x,y)
```

```python
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm, linewidth=0, antialiased=False)
```

```python
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
surf = ax.plot_surface(Fx, Fy, abs(np.fft.fftshift(np.fft.fft2(Z))*Δx*Δy)**2, cmap=cm.coolwarm, linewidth=0, antialiased=False)
```

```python
Fz = (np.sin(np.pi*(1-Fx)*w)/(np.pi*(1-Fx))-np.sin(np.pi*(1+Fx)*w)/(np.pi*(1+Fx)))*np.sin(np.pi*Fy*w)/(2*np.pi*Fy)
```

```python
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
surf = ax.plot_surface(Fx, Fy, abs(Fz)**2, cmap=cm.coolwarm, linewidth=0, antialiased=False)
```
