```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.special as sc
```

```python
λ = 6.33e-5   # Longitud de onda en cm
f = 200       # Distancia focal en cm
L = 1         # Semiperiodo espacial de la rejilla en cm
f0 = 100      # Frecuencia de la rejilla en ciclos por cm
```

```python
lim = 1
u = np.linspace(-lim, lim, 5000)
```

```python
If = (L**4/(λ**2*f**2))*(4*(np.sin(2*np.pi*L*u/(λ*f))/(2*np.pi*L*u/(λ*f)))**2+\
                         (np.sin(np.pi*L*(2*u/(λ*f)-f0))/(np.pi*L*(2*u/(λ*f)-f0)))**2+\
                         (np.sin(np.pi*L*(2*u/(λ*f)+f0))/(np.pi*L*(2*u/(λ*f)+f0)))**2)
```

```python
fig, ax = plt.subplots(figsize=(10,6))
ax.plot(u,If, color="darkgreen")
plt.xlabel(r'$u$ [cm]', fontsize=18)
plt.ylabel(r'$I_{f}$', fontsize=18)
textstr = '\n'.join((r'Ancho $=%.2f$μm' % (λ*f/L*1e4), r'Máximos en $u=0,\pm%.2f$mm' % (λ*f*f0/2*10)))
props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)
plt.text(0.05, 0.95, textstr, transform=ax.transAxes, fontsize=14, verticalalignment='top', bbox=props)
plt.savefig("../Ejercicio_37.png", dpi=150)
plt.show()
```
